```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-4, 4, 0.01)

# L1
l1 = np.abs(x)

# Clipped L1
cl1 = np.minimum(np.abs(x), 1.7)

# SCAD
lam = 0.5
a = 3.7
scad = 2 * lam * np.multiply(np.abs(x) , np.abs(x) < lam) \
        - np.multiply(np.multiply((np.power(x, 2) - 2 * a * lam * np.abs(x) + lam**2) / (a - 1), np.abs(x) < (a * lam)) , np.abs(x) > lam) \
        + (a + 1)* lam**2 * (np.abs(x) > (a * lam))

# MCP 
lam = 1
gamma =1.5
mcp = np.multiply(lam * np.abs(x) - np.power(x, 2) / (2 * gamma), np.abs(x) < (gamma * lam)) \
      + gamma * lam**2 * (np.abs(x) > (gamma * lam)) / 2
```

```python
plt.figure()
plt.plot(x, l1 , label='L1 Penalty')
plt.plot(x, cl1 , label='Clipped L1 Penalty')
plt.plot(x, scad , label='SCAD Penalty')
plt.plot(x, mcp, label='MCP Penalty')
plt.xlabel(r'$\beta$')
plt.ylabel('Penalty function value')
plt.legend()
```
